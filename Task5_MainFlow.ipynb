No. of people having or not having - Heart Disease

```python
data['target'].value_counts()
```

```python
plt.figure(figsize=(9,4))
sns.countplot(x=data['target'].value_counts().index,palette=['lightgreen','salmon'],hue=data['target'].value_counts().values)
plt.title('Heart Disease Count')
plt.show()
```

```python
plt.pie(data['target'].value_counts().values,colors=['pink','lightgreen'],shadow=True,labels=['Yes Have heart disease','Don\'t have heart disease'],autopct='%1.1f%%')
plt.title('People having Heart Disease')
plt.show()
```

APPROXIMATELY- 51% people from the given dataset have heart disease
